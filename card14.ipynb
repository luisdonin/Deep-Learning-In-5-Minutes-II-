## Etapa 1 - Operador de convulação

A convulação é o processo de adicionar cada elemento da imagem para seus vizinhos, ponderado por um kernel. Um kernel em processamento de imagem é uma pequena matrix usada para detecção de edge ou extremidades. Isso é gerado pela convolução entre o kernel e a imagem. Quando cada pixel na imagem de saída é uma função dos pixels mais próximos na imagem de entrada.

# Etapa 2 - Pooling

Reduz overfitting e ruídos desnecessários, foca nas características mais relevantes, treinado com gradient descent, com mudança de detectores de features e ajuste dos pesos.
Dois métodos comuns de pooling são média e máx pooling, 

### Max pooling em CNN
Max pooling é um tipo de operação que é tipicamente adicionada a CNN em camadas convolucionais individuais.
Quando adicionado a um modelo, max pooling reduz a dimensionalidade de imagens pela redução do número de pixels na imagem de saída da camada convolucional anterior.

### Método Average Pooling (média)
A entrada é segmentada em áreas retangulares, e uma camada de average pooling calcula a média dos valores em cada região. 

### Flattening
Após os passos anteriores, teremos um mapeamento das características, o flattening transforma esses mapas de características em uma coluna que será usado em um vetor de dados de entrada que serão passados para uma rede neural para processamento
